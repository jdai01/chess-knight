```python
from knight import Knight
import string
```

Get input from user and perform our Knight move

```python
def input_chess_notation_converter(s: str):
    """
    Function to convert to chess notation usable in our Knight class
    (as per exercise sheet input)
    """
    r = 8 - int(s[0])
    c = string.ascii_lowercase[int(s[-1])]

    return f"{c}{r}"


# Input
start_pos = input("Enter start coordinates separated by , :\n")
end_pos = input("Enter end coordinates separated by , :\n")

print(f"Start position: {start_pos}")
print(f"End position: {end_pos}")


# Tidy data to be used in my Knight class
start_pos = input_chess_notation_converter(start_pos)
end_pos = input_chess_notation_converter(end_pos)

# Load knight
k = Knight(start_pos, end_pos)
k.move()

# Output
print(f"{k.convert_board_notation(end_pos)} is reachable from {k.convert_board_notation(start_pos)} with {k.lowest_counter} moves")
```

Draw the path taken by the knight on the chessboard
- `S`: start position
- `E`: end position (and last move of knight)
- $n$: next move of knight
